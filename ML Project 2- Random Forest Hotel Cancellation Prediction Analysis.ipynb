##  Model Evalueation :- Supervised learning: Random forest Classifiaction

```python
#step1:- divide the data into x and y
```

```python
x=data.drop("is_canceled",axis=1)
y=data["is_canceled"]
```

```python
x.shape
```

```python
y.shape
```

```python
#step 2:- plot the graph 
```

```python
import matplotlib.pyplot as plt
```

```python
data.hist();
```

```python
#step 3:- split the data into x and y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
#step4:- create the model:- random forest classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RC=RandomForestClassifier(500,random_state=40)
```

```python
RC.fit(x_train,y_train)
```

```python
#step 5:-model score train and test
```

```python
print("Training Accuracy:",RC.score(x_train,y_train))
print("Testing Accuracy:",RC.score(x_test,y_test))
```

```python
#step 6:-prediction 
```

```python
yp=RC.predict(x_test)
```

```python
#step 7: accuracy score ,recall score,precision score,confusion matrix
```

```python
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix,precision_score,classification_report
```

```python
accuracy_score(yp,y_test)*100
```

```python
recall_score(yp,y_test)*100
```

```python
cm=confusion_matrix(yp,y_test)*100
```

```python
cm
```

```python
precision_score(yp,y_test)*100
```

```python
print("Classification Report:-",classification_report(yp,y_test))
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',  xticklabels=['Not Canceled','Canceled'],yticklabels=['Not Canceled','Canceled'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
```
